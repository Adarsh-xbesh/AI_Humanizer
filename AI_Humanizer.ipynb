```python
import gradio as gr

# Define the function to use the chain
def humanize_text(ai_generated_text):
    return chain.run(ai_generated_text=ai_generated_text)

# Create the Gradio interface
iface = gr.Interface(
    fn=humanize_text,
    inputs="text",
    outputs="text",
    title="Humanize Your AI Content",
    description="Enter AI-generated text to make it sound more human, engaging, and relatable.",
)

# Launch the interface
iface.launch()
```

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel

app = FastAPI()

class AIHumanizerInput(BaseModel):
    ai_generated_text: str

class AIHumanizerOutput(BaseModel):
    humanized_text: str
    human_like_score: str

@app.post("/humanize/", response_model=AIHumanizerOutput)
async def humanize_text(input: AIHumanizerInput):
    try:
        result = chain.run(ai_generated_text=input.ai_generated_text)
        print("Result from chain.run:", result)  # Debugging line
        # Assuming result is a string and needs to be parsed to extract relevant parts
        humanized_text = result.get('humanized_text', 'Error: humanized_text not found')  # Adjust according to actual result format
        human_like_score = result.get('evaluation_score', 'Error: evaluation_score not found')  # Adjust according to actual result format
        return {"humanized_text": humanized_text, "human_like_score": human_like_score}
    except Exception as e:
        print("Exception:", str(e))  # Debugging line
        raise HTTPException(status_code=500, detail=str(e))

# Example code to start the FastAPI server
if __name__ == "__main__":
    import nest_asyncio
    from pyngrok import ngrok
    import uvicorn

    # Apply the nest_asyncio patch
    nest_asyncio.apply()

    # Set your ngrok authtoken
    ngrok.set_auth_token("2k6Vj61aoDx8VhVzMIHebKpMxvT_58QZZBgRWjBv4N1dGxAC1")

    # Open a ngrok tunnel to the localhost on port 8000
    ngrok_tunnel = ngrok.connect(8000)
    print('Public URL:', ngrok_tunnel.public_url)

    # Run the FastAPI app with Uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
